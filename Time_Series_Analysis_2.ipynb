```python
# Importing required libraries
import pandas as pd
import yfinance as yf
from datetime import datetime
from datetime import date
```

```python
df_tesla =  yf.download (tickers = "TSLA", start = "2019-02-27"
                        , interval = "1D", group_by = 'ticker', auto_adjust = True, threads = True)
```

```python
df_tesla.tail()
```

## Simple Moving Average


```python

df_tesla['Open'].plot(figsize=(15,6))
```

```python
df_tesla['Open: 10 days rolling'] = df_tesla['Open'].rolling(window = 10,
                                                             min_periods=1).mean()
```

```python
df_tesla[['Open','Open: 10 days rolling']].plot(xlim=['2022-08-01','2023-08-01'],
                                                figsize=(15,6))
```

```python
df_tesla['Open: 20 days rolling'] = df_tesla['Open'].rolling(window = 20,
                                                             min_periods=1).mean()
df_tesla['Open: 30 days rolling'] = df_tesla['Open'].rolling(window = 30,
                                                             min_periods=1).mean()
df_tesla['Open: 40 days rolling'] = df_tesla['Open'].rolling(window = 40,
                                                             min_periods=1).mean()
```

```python
df_tesla[['Open','Open: 10 days rolling','Open: 20 days rolling','Open: 30 days rolling','Open: 40 days rolling']].plot(xlim=['2022-08-01','2023-08-01'],
                                                figsize=(15,6))
```
